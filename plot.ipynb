```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
history_dense = np.load('history_dense.npy', allow_pickle='TRUE')

history_rnn = np.load('history_rnn.npy', allow_pickle='TRUE')
history_rnn_pretrain = np.load('history_rnn_pretrain.npy', allow_pickle='TRUE')

history_lstm = np.load('history_lstm.npy', allow_pickle='TRUE')
history_lstm_pretrain = np.load('history_lstm_pretrain.npy', allow_pickle='TRUE')

history_cnn = np.load('history_cnn.npy', allow_pickle='TRUE')
history_cnn_pretrain = np.load('history_cnn_pretrain.npy', allow_pickle='TRUE')
```

```python
history_dense = eval(str(history_dense))

history_rnn = eval(str(history_rnn))
history_rnn_pretrain = eval(str(history_rnn_pretrain))

history_lstm = eval(str(history_lstm))
history_lstm_pretrain = eval(str(history_lstm_pretrain))

history_cnn = eval(str(history_cnn))
history_cnn_pretrain = eval(str(history_cnn_pretrain))
```

```python
history_dense
```

```python
plt.figure(figsize=(8,4),dpi=100)
epochs = np.arange(1,21)
plt.plot(epochs,history_dense["accuracy"],'bo',label='train acc')
plt.plot(epochs,history_dense["val_accuracy"],'b',marker="*",label='val acc')
plt.xlabel('epochs')
plt.ylabel('acc')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,4),dpi=100)
epochs = np.arange(1,21)
plt.plot(epochs,history_rnn["accuracy"],'bo',label='train acc')
plt.plot(epochs,history_rnn["val_accuracy"],'b',marker="*",label='val acc')
plt.plot(epochs,history_rnn_pretrain["accuracy"],'ro',label='train acc(pretraining)')
plt.plot(epochs,history_rnn_pretrain["val_accuracy"],'r',marker="*",label='val acc(pretraining)')
plt.xlabel('epochs')
plt.ylabel('acc')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,4),dpi=100)
epochs = np.arange(1,21)
plt.plot(epochs,history_lstm["accuracy"],'bo',label='train acc')
plt.plot(epochs,history_lstm["val_accuracy"],'b',marker="*",label='val acc')
plt.plot(epochs,history_lstm_pretrain["accuracy"],'ro',label='train acc(pretraining)')
plt.plot(epochs,history_lstm_pretrain["val_accuracy"],'r',marker="*",label='val acc(pretraining)')
plt.xlabel('epochs')
plt.ylabel('acc')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,4),dpi=100)
epochs = np.arange(1,21)
plt.plot(epochs,history_cnn["accuracy"],'bo',label='train acc')
plt.plot(epochs,history_cnn["val_accuracy"],'b',marker="*",label='val acc')
plt.plot(epochs,history_cnn_pretrain["accuracy"],'ro',label='train acc(pretraining)')
plt.plot(epochs,history_cnn_pretrain["val_accuracy"],'r',marker="*",label='val acc(pretraining)')
plt.xlabel('epochs')
plt.ylabel('acc')
plt.legend()
plt.show()
```
